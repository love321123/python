# 1. Agewise

```python
import pandas as pd
from matplotlib import pyplot as plt
age=pd.read_csv('AgeGroupDetails.csv')
#Data
df=age.head(9)
df
```

```python
Child=0
Teenager=0
Adult=0
Old=0
for i,value in enumerate(df.loc[:]['TotalCases']):
    if(i==0):
        Child=Child+value
    elif(i<=3):
        Teenager=Teenager+value
    elif(i<=6):
        Adult=Adult+value
    elif(i<=8):
        Old=Old+value
```

```python
plt.figure(figsize=(9,8))

labels = ['Childern','Teenager','Adult' ,'Old']
c = [Child,Teenager,Adult,Old]

plt.barh(labels,c,color='#3ACAF8',edgecolor='black',height=0.6)
plt.xlabel('No of Cases')
plt.ylabel('categorize')

plt.title('Covid-19 Cases in India (AgeWise)',fontsize=20)
plt.show()
```

# 2. Progression of Corona Virus DateWise

```python
import pandas as pd
from matplotlib import pyplot as plt
df=pd.read_excel('CovidDateWise.xlsx')
```

```python
c=df['Confirmed']
r=df['Recovered']
d=df['Deceased']
```

```python
plt.figure(figsize=(22,12))
ax = plt.axes()

plt.title('Covid-19 Cases in India (DateWise)',fontsize=24)
plt.xlabel('Date',fontsize=20)
plt.ylabel('No of cases ',fontsize=20)

date=(df['Date'])

plt.plot(date,c,label='Confirmed case', marker='o', markersize=5,markerfacecolor='black',linewidth=4)
plt.plot(date,r,label='Recovered case', marker='o', markersize=5,markerfacecolor='black',color='#49E3A6',linewidth=4)
plt.plot(date,d,label='Deceased case',marker='o', markersize=5,markerfacecolor='black',color='#DF6A6A',linewidth=4)

plt.legend(fontsize=15)
plt.grid()
plt.show()
```
